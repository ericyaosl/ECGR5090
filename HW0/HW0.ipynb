```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Question 1
# Set up all the columns for training
data = pd.read_csv('D3.csv',names = ['X1','X2','X3','Y'])
X_1 = data.values[:, 0]
X_2 = data.values[:, 1]
X_3 = data.values[:, 2]
Y   = data.values[:, 3]
m   = len(Y)
X_0 = np.ones((m,1))
```

```python
# Functions to compute cost and do gradient descent
def compute_cost(X,Y,theta):
    pred = X.dot(theta)
    errors = np.subtract(pred,Y)
    sqrErr = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErr)
    
    return J

def gradient_descent(X,Y,theta,alpha,iterations):
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        pred = X.dot(theta)
        errors = np.subtract(pred,Y)
        sum_delta = (alpha / m) * X.transpose().dot(errors);
        theta = theta - sum_delta;
        cost_history[i] = compute_cost(X,Y,theta)
    return theta, cost_history
```

```python
# X1
X_1 = X_1.reshape(m,1)
X1  = np.hstack((X_0,X_1))
theta = np.zeros(2)

# X2
X_2 = X_2.reshape(m,1)
X2  = np.hstack((X_0,X_2))

# X3
X_3 = X_3.reshape(m,1)
X3  = np.hstack((X_0,X_3))
```

```python
theta = [0.,0.]
iterations = 4000;
alpha = 0.01

theta1, cost_hist1 = gradient_descent(X1,Y,theta,alpha,iterations)
print ('Final value of theta for X1 =',theta1)
print ('cost history for traning X1 =',cost_hist1)

theta2, cost_hist2 = gradient_descent(X2,Y,theta,alpha,iterations)
print ('Final value of theta for X2 =',theta2)
print ('cost history for traning X2 =',cost_hist2)

theta3, cost_hist3 = gradient_descent(X3,Y,theta,alpha,iterations)
print ('Final value of theta for X3 =',theta3)
print ('cost history for traning X3 =',cost_hist3)
```

```python
plt.scatter(X1[:,1],Y,color='red',marker='+',label='Training Data')
plt.plot(X1[:,1],X1.dot(theta1),color='green',label='Linear Regression')

plt.rcParams["figure.figsize"]=(10,6)
plt.grid()
plt.xlabel('X1 = Column 1')
plt.ylabel('Y = Column 4')
plt.title('Linear Regression X1')
plt.legend()
```

```python
plt.scatter(X2[:,1],Y,color='red',marker='+',label='Training Data')
plt.plot(X2[:,1],X2.dot(theta2),color='green',label='Linear Regression')

plt.rcParams["figure.figsize"]=(10,6)
plt.grid()
plt.xlabel('X2 = Column 2')
plt.ylabel('Y = Column 4')
plt.title('Linear Regression X2')
plt.legend()
```

```python
plt.scatter(X3[:,1],Y,color='red',marker='+',label='Training Data')
plt.plot(X3[:,1],X3.dot(theta3),color='green',label='Linear Regression')

plt.rcParams["figure.figsize"]=(10,6)
plt.grid()
plt.xlabel('X3 = Column 3')
plt.ylabel('Y = Column 4')
plt.title('Linear Regression X3')
plt.legend()
```

```python
plt.plot(range(1,iterations+1),cost_hist1,color='blue',label='X1 Cost')
plt.plot(range(1,iterations+1),cost_hist2,color='green',label='X2 Cost')
plt.plot(range(1,iterations+1),cost_hist3,color='red',label='X3 Cost')
plt.rcParams["figure.figsize"]=(10,6)
plt.grid()
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')
```

```python
# Question 2
# Set up a m x 4 array for X5
X5  = np.hstack((X_0,X_1,X_2,X_3))
theta5 = np.zeros(4)

theta5 = [0.,0.,0.,0.]
iterations5 = 4000;
alpha5 = 0.01;

theta5, cost_hist5 = gradient_descent(X5,Y,theta5,alpha5,iterations5)
print ('Final value of theta for Q2 =',theta5)
print ('cost history for traning Q2 =',cost_hist5)
```
